```python
data = pd.read_excel(r"C:\Users\ADMIN\Desktop\Data Science\LetsUpgrade\Datasets\Bank_Personal_Loan_Modelling.xlsx", sheet_name=1)
data.head()
```

```python
data = data.drop(columns = ['ID', 'ZIP Code'] )
data.drop_duplicates(inplace=True)
data.shape
```

```python
Y = data['Personal Loan']
X = data[['Age', 'Experience', 'Income', 'Family', 'CCAvg', 'Education', 'Mortgage', 'Securities Account', 'CD Account', 'Online', 'Loan Approval']]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_cv, y_train, y_cv = train_test_split(X,Y, test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
model = LogisticRegression()
model.fit(x_train, y_train)
LogisticRegression()
```

```python
pred_cv = model.predict(x_cv)
print("accuracy percentage is: ",accuracy_score(y_cv,pred_cv)*100)
```
